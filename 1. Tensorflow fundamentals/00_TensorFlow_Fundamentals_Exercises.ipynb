### 1. Create a vector, scalar, matrix and tensor with values of your choosing using tf.constant().


```python
import tensorflow as tf

scalar = tf.constant(10)
scalar
```

```python
vector = tf.constant([7, 10])
vector
```

```python
matrix = tf.constant([[10, 7], 
                      [7, 10]])

matrix
```

### 2. Find the shape, rank and size of the tensors you created in 1.


```python
print(f"Shape of the scalar is: {scalar.shape}\n Rank of the scalar is: {tf.rank(scalar)}\n The size of the scalar is: {tf.size(scalar)}")
```

```python
vector.shape, tf.rank(vector), tf.size(vector)
```

```python
matrix.shape, tf.rank(matrix), tf.size(matrix)
```

### 3. Create two tensors containing random values between 0 and 1 with shape [5, 300].


```python
t1 = tf.random.uniform(shape=(5, 300))
t2 = tf.random.uniform(shape=(5,300))
```

```python
t1
```

```python
t2
```

4. Multiply the two tensors you created in 3 using **matrix multiplication**.


```python
tf.multiply(t1, t2)
```

5. Multiply the two tensors you created in 3 using **dot product**

```python
tf.tensordot(t1, t2, axes=0)
```

6. Create a tensor with random values between 0 and 1 with shape [224, 224, 3].


```python
t6 = tf.random.uniform((224, 224, 3))
t6
```

7. Find the **min** and **max** values of the tensor you created in 6.


```python
print(f"Minimum value of the tensor: {tf.reduce_min(t6)}")
print(f"Max value of the tensor: {tf.reduce_max(t6)}")
```

8. Created a tensor with random values of shape [1, 224, 224, 3] then squeeze it to change the shape to [224, 224, 3].


```python
t8 = tf.random.uniform(shape=(1, 224, 224, 3))
t8.shape
```

```python
t8_squeezed = tf.squeeze(t8)
t8_squeezed.shape
```

9. Create a tensor with shape [10] using your own choice of values, then find the index which has the maximum value.


```python
t9 = tf.constant([1, 2, 3, 4, 0.5, 6, 7, 8, 9, 0])
t9
```

```python
print(f"Index having max value: {tf.argmax(t9)}\nIndex having min value: {tf.argmin(t9)}")
```

10. One-hot encode the tensor you created in 9.


```python
t9_cast = tf.cast(t9, dtype=tf.int32)
print(f"Original tensor: {t9}\nCasted tensor: {t9_cast}")
```

```python
tf.one_hot(t9_cast, depth=len(t9_cast))
```
